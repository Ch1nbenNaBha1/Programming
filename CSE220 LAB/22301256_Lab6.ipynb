```python
#Task 1

#1.a Recursive algorithm to find factorial of n

def fact(n):
    if n ==1:
        return 1
    return n+ fact(n-1)

#1.b recursive algorithm to find the n-th Fibonacci number
def fibo(n):
    if n == 1 or n == 2:
        return 1
    return fibo(n-1) + fibo(n-2)

#1.c Print all the elements of a given array recursively.

def printArray(arr, n = 0):
    if n < len(arr):
        return
    print(arr[n-1])
    printArray(arr, n-1)

#1.d algorith to compute power of a number using recursion

def pow(n , power = 0):
    if power == 0:
        return 1
    return n * pow(n, power-1)
```

```python
# Test for task 1
#a
print(fact(5))
#b
print(fibo(5))
#c
printArray([1,2,3,4,5])
#d
print(pow(2,3))
```

```python
#Task 2

#2.a convert n to binary

def binary(n):

    if n==0:
        return 0
    else:
        return n%2+10*binary(n//2)
    
#2.b add all the elements of linked list

def addElem(head):
    if head==None:
        return 0
    else:
        return head.data+addElem(head.next)
    
#2.c print the linked list in reverse order

def printReverse(head):
    if head==None:
        return
    else:
        printReverse(head.next)
        print(head.data)
```
