```python
class HashTable():
    def __init__ (self, size):
        self.size = size
        self.slots = [None] * self.size
        self.data = [None] * self.size

    # Method for putting a key-value pair in the hash table
    def inset(self, key, data):
        # Get the hash value of the key
        hashValue = self.hashFunction(key, len(self.slots))

        # If the slot is empty, insert the key and data
        if self.slots[hashValue] == None:
            self.slots[hashValue] = key
            self.data[hashValue] = data
        # If the slot is not empty, check if the key is the same
        # If the key is the same, replace the data
        # If the key is not the same, find the next available slot
        else:
            if self.slots[hashValue] == key:
                self.data[hashValue] = data
            else:
                nextSlot = self.rehash(hashValue, len(self.slots))
                while self.slots[nextSlot] != None and self.slots[nextSlot] != key:
                    nextSlot = self.rehash(nextSlot, len(self.slots))
                if self.slots[nextSlot] == None:
                    self.slots[nextSlot] = key
                    self.data[nextSlot] = data
                else:
                    self.data[nextSlot] = data

    # Method for getting the hash value of a key
    def hashFunction(self, key, size):
        return key % size
    
    # Method for rehashing
    def rehash(self, oldHash, size):
        return (oldHash + 1) % size
    
    # Method for getting the value of a key
    def get(self, key):
        startSlot = self.hashFunction(key, len(self.slots))
        data = None
        stop = False
        found = False
        position = startSlot

        while self.slots[position] != None and not found and not stop:
            if self.slots[position] == key:
                found = True
                data = self.data[position]
            else:
                position = self.rehash(position, len(self.slots))
                if position == startSlot:
                    stop = True
        return data
    
    # Method for getting the value of a key
    def __getitem__(self, key):
        return self.get(key)
    
    # Method for putting a key-value pair in the hash table
    def __setitem__(self, key, data):
        self.inset(key, data)

# Test the hash table
H = HashTable(5)

H.inset(0, 'zero')
H.inset(1, 'one')
H.inset(2, 'two')
H.inset(3, 'three')
H.inset(4, 'four')
H.inset(5, 'five')
H.inset(6, 'six')
H.inset(7, 'seven')
H.inset(8, 'eight')
H.inset(9, 'nine')
H.inset(10, 'ten')

print(H.slots)
print(H.data)


```
