1. You need to find the mean of a series of numbers stored in an array.
2. You need to calculate the sample standard deviation of those numbers.
3. You need to create a new array where you will only store the numbers that are at least 1.5 standard deviations away from the mean.

```python
# LAB1 Part 2
#1
def findMean(array):
  sum = 0
  for i in range(len(array)):
    sum += array[i]
  return sum/len(array)
#2
def findStd(array):
  mean = findMean(array)
  sum = 0
  for i in range(len(array)):
    sum += (array[i]-mean)**2
  return (sum/(len(array)-1))**0.5
#3
def findOutliers(array):
    mean = findMean(array)
    std = findStd(array)
    count = 0
    for i in range(len(array)):
        if abs(array[i]-mean) > 1.5*std:
            count += 1
    outliers = np.zeros(count)

    j = 0
    for i in range(len(array)):
        if abs(array[i]-mean) > 1.5*std:
            outliers[j] = array[i]
            j += 1
    
    return outliers

sample = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])

print("///  Test 07: Find Mean  ///")
returned_value = findMean(sample)
print(f'Task 7: {returned_value}')

print("///  Test 08: Find Standard Deviation  ///")
returned_value = findStd(sample)
print(f'Task 8: {returned_value}')

print("///  Test 09: Find Outliers  ///")
returned_value = findOutliers(sample)
print(f'Task 9: {returned_value}') 

```
