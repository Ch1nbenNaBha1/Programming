```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  print('-'*(col*6)+'-')
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

```python
#DO NOT CHANGE THE PARAMETERS OR THE RETURN STATEMENT OF THE FUNCTION
def moving_around(wrld,cmds):
    # TO DO
    start_x = 3 
    start_y = 3
    wrld[start_x][start_y] = '@'
    current_x = 3
    current_y = 3
    for i in range(len(cmds)):
        if cmds[i] == "U":
            wrld[current_x-1][current_y] = '*'
            current_x -= 1
        elif cmds[i] == "D":
            wrld[current_x+1][current_y] = '*'
            current_x += 1
        elif cmds[i] == "L":
            wrld[current_x][current_y-1] = '*'
            current_y -= 1
        elif cmds[i] == "R":
            wrld[current_x][current_y+1] = '*'
            current_y += 1
            
    wrld[current_x][current_y] = '#'
        
    return wrld


#DO NOT CHANGE ANYTHING BELOW
#The grid has already been generated for you.
cmd1 = np.array(['U','U','R','D','D','D','D','L'])
cmd2 = np.array(['D','D','L','L','U'])
cmd3 = np.array(['U','R','R','U','L'])
cmd4 = np.array(['R','D','R','D','D','L','L','U','L','L','U','U','U'])
for i,commands in enumerate((cmd1,cmd2,cmd3,cmd4),start=1):
    world = np.array([['.']*7]*7)
    world2D = moving_around(world,commands)
    print('='*18+f"TEST#{i}"+'='*18+'=')
    print_matrix(world2D)
```
